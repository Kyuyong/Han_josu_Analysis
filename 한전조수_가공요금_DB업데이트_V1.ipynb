```python
## Path 설정
강원path = 'd:/Dev/한전조수분석/data/'+month+'/강원본부/'
경기path = 'd:/Dev/한전조수분석/data/'+month+'/경기본부/'
경기북부path = 'd:/Dev/한전조수분석/data/'+month+'/경기북부본부/'
경남path = 'd:/Dev/한전조수분석/data/'+month+'/경남본부/'
경북path = 'd:/Dev/한전조수분석/data/'+month+'/경북본부/'
남서울path = 'd:/Dev/한전조수분석/data/'+month+'/남서울본부/'
부산path = 'd:/Dev/한전조수분석/data/'+month+'/부산본부/'
서울path = 'd:/Dev/한전조수분석/data/'+month+'/서울본부/'
인천path = 'd:/Dev/한전조수분석/data/'+month+'/인천본부/'
전남path = 'd:/Dev/한전조수분석/data/'+month+'/전남본부/'
전북path = 'd:/Dev/한전조수분석/data/'+month+'/전북본부/조수별 공가요금 세부내역 조회/'
제주path = 'd:/Dev/한전조수분석/data/'+month+'/제주본부/'
충남path = 'd:/Dev/한전조수분석/data/'+month+'/충남본부/ST000가공/'
충북path = 'd:/Dev/한전조수분석/data/'+month+'/충북본부/'

## 전남 Path는 불필요한 파일이 많아서 별도의 Logic이 필요함 (조수별이 등록된 파일만 읽어오기)
def 조수별데이터불러오기(path) : 
    대상 = os.listdir(path)
    search = '조수별'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr')
        else :
            pass
    return df

def 강원데이터불러오기(path) : 
    대상 = os.listdir(path)
    search = '가공'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr')
        else :
            pass
    return df

def 경기북부데이터불러오기1(path) : 
    대상 = os.listdir(path)
    search = 'ST000.xls'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr', sheet_name='조수별 공가요금 세부내역1')
        else :
            pass
    return df


def 경기북부데이터불러오기2(path) : 
    대상 = os.listdir(path)
    search = 'ST000.xls'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr', sheet_name='조수별 공가요금 세부내역2')
        else :
            pass
    return df


def 경북데이터불러오기(path) : 
    대상 = os.listdir(path)
    search = 'SK계열'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr')
        else :
            pass
    return df

def 부산데이터불러오기(path) : 
    대상 = os.listdir(path)
    search = '세부내역'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr', sheet_name='가공세부내역')
        else :
            pass
    return df

def 충북데이터불러오기(path) : 
    대상 = os.listdir(path)
    search = '가공요금'
    for i in range(0,len(대상)):
        if search in 대상[i]:
            df = pd.read_excel(path+대상[i], encoding = 'euc-kr')
        else :
            pass
    return df

# splite 파일 불러오기
def open_sqlite(sqlite_file, sql_str):
    conn = sqlite3.connect(sqlite_file)
    ret = pd.read_sql(sql_str, conn)
    conn.close()
    return ret 

def save_sqlite(sqlite_file, table_name ,df):
    conn = sqlite3.connect(sqlite_file)
    df.to_sql(table_name, conn,if_exists='append')
    conn.close()
```

```python
# splite 파일 불러오기
def open_sqlite(sqlite_file, sql_str):
    conn = sqlite3.connect(sqlite_file)
    ret = pd.read_sql(sql_str, conn)
    conn.close()
    return ret 


sqlite = 'd:/dev/SQL_DB/한전조수DB.db'
sql_str = 'SELECT * FROM 사업소코드'

사업소코드 = open_sqlite(sqlite, sql_str)
사업소코드 = 사업소코드.drop(columns=['index'])
```
